# Função Distribuição Acumulada (fda em português ou cdf em inglês)

Representa a probabilidade de uma variável aleatória pertencer a certos intervalos.
Para cada ponto $(x, F(x))$ da fda:

$F(x)=P(X<x)=\text{A probilidade da variável aleatória X ser menor do que x}$



Baseado em $F(x)$, podemos calcular $P(a < x < b) = F(b) - F(a)$

A fda pode ser calculada a partir da Função Densidade de Probabilidade (fdp) da variável aleatória X, $f(x)$:

$F(x) = \int_{0}^{\infty} f(x)*dx$
