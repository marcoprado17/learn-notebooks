# Função Densidade de Probabilidade (aka fdp)

Representa a probabilidade relativa de uma variável aleatória X assumir um determinado valor.



$P(a < X < b) = \int_{a}^{b} f(x)dx$

Se uma variável aleatória tem densidade dada por $f(x)$, então o intervalo infinitesimal $[x, x+dx]$
tem probabilidade $f(x) dx$
