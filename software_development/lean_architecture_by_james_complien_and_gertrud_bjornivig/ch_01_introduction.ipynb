# 1. Introduction

Esse capítulo foca em explicar a filosofia e a forma como o autor irá expor as ideias nos capítulos seguintes.

## 1.1 The touchstones: Lean and Agile

* Lean e Agile estão em voga atualmente, lean remete ao minimalismo (mas não somente ao minimalismo). Lean possui raizes no Japão, mais especificamente na montadora Toyota;
* No Lean, o minimalismo é consequência de uma abordagem na qual a visão de criação de valor PARA TODOS OS STAKEHOLDERS (empregado, chefe, cliente, parceiro etc) permeia a todo tempo, em todas as dimensões. Lean não é apenas suas práticas (just in time, end to end). No manifesto Agile, a enfase é mais o cliente;
* Lean e agile não são sobre hard work ou smart work, mas sim sobre FOCO e DISCIPLINA no cumprimento de ideis simples;
* O livro foca em aplicar o estilo Lean e Agile para o a arquitetura de software;
* Arquitetura de software por ser definida como os fundamentos que guiam a organização dos componentes do sistema, a relações destes, o ambiente e os principios que guiam seu desenvolmento e evolução;
* As práticas do Lean são importantes, mas no seu core está a participação de todos do time, desde o início do projeto. Isso pode ser meio caótico no inicio, mas segundo o autor vale a pena;
* Arquitetura pode paracer dificil, mas não é, basta reunir os especialistas em uma sala. Individuos vem primeiro, interações em segundo plano;
* O Lean contribui com o planejamento com TODOS os stakeholders com foco na criação de valor e o Agile contribui para a cultura do feedback;
* Como o Lean foca na criação de valor para o cliente, ele contribui para a redução de software que não entrega valor para o cliente. 70% do software contruído não é utilizado pelo cliente;
* Uma Arquitetura Lean é capaz de produzir código que entregue valor para o cliente e permitir adições futuras sem q para isso tenhamos que criar código de ante-mão. Foca no que é importante agora, sem prejudicar o longo termo;
* Um arquitetura Lean permite que os principios agéis ocorram com sucesso, é a base para o desenvolvimento ágil, bem sucedido:
    * All hands on desk => Individuals and interactions over processes and tools;
    * Value stream without many processes and tools => Customer engagement;
    * Software that reflect the end user mental model => Working software;
    * Software that captures domain form => Software that can respond to change.






## 1.2 Lean Architecture and Agile Feature Development

* A Arquitetura Lean une a visão da Toyota à visão Agile,o Lean contribui para a form de como o sistema será desenvolvido focando na experiência e no dominio e a perpectiva Agile foca na resposta e planejamento às mudanças;
* Diferencas entre a Arquitetura Lean (a esquerda) e a arquitetura clássica (a direita) e obserações a respeito dessas diferenças:
    * Adia engenharia x Inclui engenharia. No Lean, a forma deve vir da UX, eventualmente as leis que governam a engenharia virão, mas não devem ser as primeiras; 
    * Espaço para mudanças x Limita mudanças bruscas. O Lean foca na forma do dominio, forma e funcionalidade sem prematura implementação, de forma que não ocorrem mudanças, apenas extensões;
    * Adia implementações (entrega API's e descrição das relações) x Inclui muita implementação ou apenas documentação. Não precisamos implementar ou documentar tudo logo de cara para buscar implementação de padrões, o Lean valoriza os padrões, mas esse não precisam ser estabelecidos na implementação, mas nas API's, na forma, nos protocolos etc;
    * Doumentação leve x Documentação focada em descrever a implementação. No Lean, muitas decisões podem ser tomadas no código, ao focar nas pessoas, muitas das decisões podem ser estabelecidas no face to face. É claro q no Lean, podem existir documentações se necessário;
    * Pessoas x Ferramentas e Notações. O Lean valoriza mais as pessoas e seus experties em suas respectivas àreas;
    * Planejmanento conjunto e cooperação x Planejamento especializado e controle. Tanto no Lean qto no clássico possuem visão de longo prazo, mas o Lean não segue o planejamento a risca, ele foca no que é importante agora, seja fazer a etrega para a próxima semana, seja fazer o planejamento de longo prazo;
    * Modelagem com base no usuário final x Acoplamento técnico e coesão;

## 1.3 Agile Production

Um design Lean leva a uma arquitetura que te permite ser mais Agile segundo o manifesto:

> We are uncovering better ways of developing
> software by doing it and helping others do it.
> Through this work we have come to value:

> * *Individuals and interactions* over processes and tools
> * *Working software* over comprehensive documentation
> * *Customer collaboration* over contract negotiation
> * *Responding to change* over following a plan

> That is, while there is value in the items on
> the right, we value the items on the left more.

### 1.3.1 Agile Build on Lean

* Agile se prepara para mudanças na forma como o sistema deve se comportar, o Lean abraça a mudança a partir da fato que o Lean se baseia em uma corrente continua de adição de valores pelos stakeholders, o que fez sentido no passado, não necessáriamente faz sentido agora;
* Agile foca em código funcionando, o Lean supoe que a melhor forma de expressar o design e a modelagem mental do usuário é por meio do código. É claro que o Lean concorda que documentações leves podem eventualmente ser necessárias para os stakeholders não programadores;
* Arquitetura classíca foca naquilo q não muda, Agile foca naquilo que é mais provável de mudar (TOAGREE). A arquitetura Lean é um intermediário entre essas duas formas;
* Lean foca na cadeia de valores começando a partir do usuário final, o que está de acordo com a abordagem Agile focada em pessoas e iterações com foco no cliente;
* A redução de desperdicio com o Lean está de acordo com a redução de documentação desnecessária do Agile;
* Lean e Agile não são conceitos opostos, são perspectivas diferentes para construção de sistemas com foco em produtividade e qualidade;

### 1.3.2 The Scope of Agile Systems

* Nos sistemas atuais em muitos mercados (web, financial etc) as mudanças ocorrem em termos de segundos. O usuário interage com o sistema e estabelece qual a input momentos antes de precisar da resposta, não há mais aquela ideia de se preparar um lote de inputs a serem proecessadas por um longo periodo de tempo;
* A metodologia ágil comporta bem um ambiente de rápidas mudanças. Porêm o Agile não é muito útil em ambientes onde o feedback é pouco valorizado ou há tantas restrições (seja por tecnologias ou seja por ser necessário seguir um protocolo) que não iterações entre pessoas;
* Lean brilha mais em áreas específicas, para problemas simples ele é matador, mas para problemas mais complexos o agile pode contribuir para o Lean;
* Um sistema pode possuir partes que são melhor comportadas pelo Lean e outras partes q são melhor comportadas pelo Agile. Em uma arquitetura de software em camadas, a camada do dominio é beneficiada por domain expertise e Lean análise e parte de interação com o usuário é beneficiada pelo Agile;

### 1.3.3 Agile and DCI

* DCI promete capturar no código o modelo mental do usuário, e nesse caso, as chances do código funcionar (no sentido de fazer o que o usuário espera) são maior;

## 1.4 The Book in a very small Nutshell

* Um sistema deve refletir o modelos mental do usuário. Há a representação do processo mental do usuário ao ver a tela, a forma do sistema, e também como o sistema responde a interação com o usuário, a funcionalidade. Nos devemos capturar a forma e as funcionalidades o mais antes possível, Acomplamento e coesão são secundários;
    * Coesão de um sistema representa se as partes de uma classe ou módulo estão alinhados com um mesmo objetivo que não seja muito genérico
    * Acomplamento de um sistema mede o grau de dependencias entre seus modulos e classes. Se um módulo ou classe muda, o quanto isso impacta outros módulos e classes
    * Um bom software possui alta coesão e baixo acoplamento.
* Para explorar forma e funcionalidade todos os stakeholders devem estar envolvidos desde o ínicio, isso torna o processo de desenvolvimento mais efetivo;
* Lean prioriza forma (abstrações e contratos) ante estrutura (implementação);
* Podemos expressar funcionalidades do sistema por meio de use cases;
* Podemos traduzir use cases em algorithmos just in time, conforme eles vão surgindo. As classes capturam a forma/dominio e o roles (construídos a partir dos use cases) capturam funcionalidades. Separar forma e funcionalidade possui profunda implicações em compreensão do código, testabilidade e análise formal;
* É utilizado o conceito de trait (conceito de OO semelhante ao interface, porêm pode conter implementação) para permitir a conexão entre uma classe de domínio e uma classe de role;

## 1.5 Lean and Agile: Contrasting and Complementary

* Cascata foca muito no planejamento, mas não no que acontece depois da entrega em como o sistema muda após sua entrega. Agile, em contrapartida, foca muito na mudança e pouco no planejamento. Zero planejamento pode prejudicar a construção de sistemas lucrativos e com certos padrões;
* Agile e Lean nasceram para resolver problemas, há algumas similaridades e distinções
* Arquitetos criam documentações que logo se tornam obseletas, o Agile remove a necessidade dessas documentações (documentação no próprio código). A forma correta de se enxergar software é que depois da primeira compilação, tudo é manutenção;
* Os conceitos de Lean e Scrum são separados e complementares, surgindo a partir da observação da construção de sistemas complexos e adaptativos;
* Comparação entre Lean (a esquerda) e Agile (a direita):
    * Thinking and doing x Doing;
    * Inspect-plan-do x Do-inspect-adapt;
    * Feed-forward and feedback (design for change and react to change) x Feedback (react to change);
    * High throughput x Low latency;
    * Planning and responsing x Reacting;
    * Focus on process x Focus on people;
    * Teams (working as unit) x Individuals (and interactions);
    * Complicated systems x Complex systems;
    * Embrace standards x Inspect and adapt;
    * Reworking in design adds value, in making is waste x Minimize up-front work of any kind and rework code to get quality;
    * Bring descisions forward (Decision Structure Matrices) x Defer decisions (to the last responsible moment).
* O Lean é mais indicado para sistemas previsíveis e estáveis (know knows predominam), mas que podem ser complicados, nesse caso vale a pena investir tempo planejando antes de fazer;
* O Agile é mais indicado para sistemas complexos (unknow unknows e know unknows predominam), nesse caso fazer algo que funcione e experimentar tem mais valor do que planejar;
* O fato de um sistema ser complexo (unknow unknows e know unknows predominam) ou simplesmente complicado (know knows predominam) está relacionado às pessoas, um sistema X pode ser complexo para João, mas apenas complicado para Maria (especiliasta no dominio do sistema);
* A arquitetura de um sistema, tende a ser estável pois custa muito muda-lá, de modo que uma abordagem Lean para a arquitetura de um sistema faz mais sentido do que uma abordagem ágil. É uma arquitetura Lean que suporta uma abordagem Agile eficiente para projetos demorados;

## 1.6 Práticas perdidas

O foco do livro está em prover aquilo que a abordagem Agile do XP ou Scrum não é tão eficaz ou lembrado: arquitetura de software e sistemas, requiremens dependency management, foundations for usability, documentation, and others;

### 1.6.1 Architecture

Um projeto deve ser forte para suportar mudança. Essa força pode vir de sua arquitetura. Investir em arquitetura no inicio vale a pena no longo termo e o preço a se pagar por investir em arquitetura não precisa ser alto, ele será alto se alêm em nossa arquitetura investimro na estrutura/implementação. Na etapa de arquitetar o sistema não precisamos da estrutura/implementação, precisamos da forma/classes abstratas e interfaces

### 1.6.2 Handling Dependencies between requirements

A ausência do cliente leva a criação de novos requisitos incosistentes com os requisitos anteriores

### 1.6.3 Foundations for Usability

Agile fala sobre "working software", mas não sobre "usable software", bons sistemas capturam o modelo mental do usuário da forma mais fidedigna possível, o trabalho de Trygve de MVC + DCI foca exatamente nisso.

### 1.6.4 Documentation

* Documentação possui uma má reputação, muitos esquecem de sua importância: comunicar perspectivas e decisões, seja como lembrete para os que participaram, seja como tutorial para os novos integrantes;
* Se a decisão e perspectiva pode ser observada no código, não há necessidade de haver documentação duplicada;
* Não devemos enxergar a documentação como algo a ser mantido, se manter atualizado, mas sim um refletir o design / decisão tomado naqeuele momento. A documentação não necessariamente é um documento em um certo padrão, pode ser simplesmente um papel com anotações (comunicação menos rica e eficiente) até uma discussão de dois engenheiros com o auxílio de um quadro branco (comunicação mais rica e eficiente);
* A documentação no Lean não deve ser pesada, pois conseguimos expressar boa parte das decisões comunicando-as por meio do código;
* Emboara o código seja em si o proprio design, ele não consegue expressar tudo, o código consegue expressar o "what" e o "how", mas não o why (GoF);
* Embora o código represente os design, talvez seja mas simples visualizar a documetação com uma visão mais ample, utilzar ferramentas que gerem documentação a partir do código é um boa;




## 1.7 What This Book is Not About

* O livro não possui receitas para uso, o que importa são as ideias, não o processo;
* Embora o livro fale muito que o software não é um produto mas sim um serviço, o livro não trata sobre Service Oriented Architecture
* Dado que já existe outras fontes mais propicias, o livro não trata de assuntos como concorrência, distruibuição e segurança;
* O livro não trata sobre performance. Otimização prematura é a origem de todo o mal. A melhor forma de melhor a performance do sistema é descobrir as partes que tomam mais tempo e otimizá-las;
* O livro não trata sobre design;

## 1.8 Agile, Lean - Oh, Yeah and Scrum and Methodologies and Such

* Scrum e XP estão em moda. Não são o mesmo que Agile, muito menos Lean. Mas o Scrum captura boa parte do Lean, do Agile e novas coisas;
* O livro é sobre Lean aplicado á arquitetura, embora existam o XP, o scrum etc, todos remontam a aplicação do Lean inicialmente utilizada pela Toyota;
* Quando o autor se refere a Agile ele está se referindo ao manifesto, que traz agilidade por consequência, mas é muito mais focado em self-organization e feedback;
* Scrum é uma framework agile que busca aumentar o ROI utilizado:
    * As coisas e caminhos mais importantes são feitos primeiro
    * Ciclos curtos de entrega;
    * Desenvolvededores possuem informações de negócios para tomar decisões;
* Scrum não é especifico para nenhum tipo de negócio, podemos utiliza-lo tanto em uma indústria quanto no desenvolvimento de software;
* Abordagens Agile, incluindo o Scrum, são baseadas em três pilares:
    * Trust;
    * Communication;
    * Self organization.
